```python
from glob import glob
# import glob는 에러난다. 
# module 을 함수처럼 사용하려 (괄호)를 통해 불러오면 뜨는 에러
# https://blog.naver.com/PostView.naver?blogId=shj961105&logNo=222476213070
import re


folder_label = 59   # 기존의 라벨에 이은 FOOD Index 60번 부터
for i in ver2:
    ver2_path = path + i + '/'
    file_list = os.listdir(ver2_path)
    print(file_list)
    for last_folder in file_list:
#         print(last_folder)
        folder_label += 1
        file_path = glob(ver2_path + last_folder + '/*')
#         print(file_path)

############# "CU-찜갈비도시락"을 "CU_찜갈비도시락"으로 바꾸기 ############
        p = re.compile('[-]')
        _last_folder = p.sub('_', last_folder)
###########################################################################
        
    
        file_name_list = []
        for num, original_path in enumerate(file_path, start=1):
#             print(num, original_path)   # 새 이름에 들어갈 생성번호, 기존 이름의 path
            file_name = 'dki'  + '_' + _last_folder + '_' + str(folder_label) + '_9{0:05d}'.format(num) + '.jpg'   # 새 이름 
            file_name_list.append(file_name)
            rename_path = ver2_path + last_folder + '/' + file_name
#             print(rename_path)

            # 얘 실행시키면 이름 다 바뀐다.
            os.rename(original_path, rename_path)

#         print(file_name_list[0])
#         print(file_name_list[-1])
    
```
