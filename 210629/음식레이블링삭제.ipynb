```python
import pandas as pd
import numpy as np

# !pwd

df = xlsx = pd.read_excel('가공식품리스트_20210628.xlsx')
df = df[['FOOD_CD', 'FOOD_NM']].set_index('FOOD_CD')
df
```

```python
'''

후에 jpg 파일 이름과 FOOD_NM 을 정규표현식으로 매칭할 때를 위해서
정규표현식을 사용해서 [,],(,)는 앞에 \를 붙여준다.

'''


import re
p1 = re.compile(r'(\[)')
p2 = re.compile(r'(\])')
p3 = re.compile(r'(\()')
p4 = re.compile(r'(\))')

def change(value):
  value = p1.sub('\[', value)
  value = p2.sub('\]', value)
  value = p3.sub('\(', value)
  value = p4.sub('\)', value)
  return value
df.FOOD_NM = df.FOOD_NM.apply(lambda x : change(x))
df
```

```python
'''

FOOD_CD를 키, FOOD_NM를 벨류로 딕셔너리를 만든다.

'''

df_dict = df.to_dict()['FOOD_NM']
df_dict
```
