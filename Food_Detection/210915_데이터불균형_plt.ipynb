```python
import os

os.chdir('/content/drive/MyDrive/음식인식/data/packed_2_삭제')
os.getcwd()
```

```python
path_dir = '/content/drive/MyDrive/음식인식/data/packed_2_삭제'

folders = os.listdir()
len(folders), folders
```

```python
bin = []

for folder in folders:
  change_dir = os.path.join(path_dir, folder)
  os.chdir(change_dir)
  bin.append(len(os.listdir()))

print(bin)
```

```python
import pandas as pd
import numpy as np

df_make = pd.DataFrame({'name':folders, 'num':bin})
df_make
```

```python
os.chdir('/content/drive/MyDrive/음식인식/')
os.getcwd()

df_name = pd.read_excel('가공식품_이종희.xlsx')[['FOOD_CD', 'FOOD_NM', 'Unnamed: 2']].rename(columns={'Unnamed: 2':'contents'})
df_name
```

```python
for i in range(len(df_make)):
  for j in range(len(df_name)):
    if df_make.name[i] == df_name.FOOD_CD[j]:
      df_make.name[i] = df_name.FOOD_NM[j]

df_make
```
