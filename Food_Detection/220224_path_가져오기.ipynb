```python
def 폴더별로파일하니씩뽑아서csv로저장(path, folder):
    
    index = 0
    df = pd.DataFrame(columns = ['folder_path', 'jpg_path', 'txt_path'])
    
    path = os.path.join(path, folder)
    for (path, dir_, files) in os.walk(path):
        if files:
            #print(files[0])
            jpg_path, txt_path ='', ''
            확장자 = os.path.splitext(files[0])[-1]
            #print(확장자)
            if 확장자 == '.jpg':
                jpg_path = os.path.join(path, files[0])
                txt_path = jpg_path[:-3]+'txt'
                #print('path: ', path)
                #print('jpg_path: ', jpg_path)
                #print('txt_path: ', txt_path)
                #print('='*50)
            
            elif 확장자 == '.txt':
                txt_path = os.path.join(path, files[0])
                jpg_path = txt_path[:-3]+'jpg'
                #print('path: ', path)
                #print('jpg_path: ', jpg_path)
                #print('txt_path: ', txt_path)
                #print('='*50)
            
            df.loc[index] = [path, jpg_path, txt_path]
        index += 1

    # utf-8은 한글이 깨져서 euc-kr로 해라!
    df.to_csv(f'./{folder}.csv', encoding='euc-kr')
    print(f'{folder}.csv 생성 완료!')
```

```python
for folder in folders:
    폴더별로파일하니씩뽑아서csv로저장(os.getcwd(), folder)
```
