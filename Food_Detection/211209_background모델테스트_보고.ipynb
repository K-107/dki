# 웹크롤링 데이터 훈련 모델 평가

```python
import os
import shutil
import glob
from glob import glob
from PIL import Image
import re

import pandas as pd
import numpy as np
```

### 웹크롤링 데이터 통계

```python
val_img_path = '/data/Data_data/FoodDetection/data/inspection/도시락/도시락'
os.listdir(val_img_path)
```

```python
df = pd.DataFrame(columns=['class_name', 'class_num','count'])

for img in os.listdir(val_img_path):
    
    p_name = re.compile(r'.*(?=[_]\d*[_])')
    p_num = re.compile(r'[_]\d*[_]')
    
    m_name = p_name.search(img)
    m_num = p_num.search(img)
    
    name = m_name.group()
    num = m_num.group()[1:-1]
    
    df = df.append({'class_name':name,'class_num':num,'count':1}, ignore_index=True)
df = df.sort_values(by='class_num', ascending=True)
df = df.reset_index()
df.drop('index', axis=1, inplace=True)
df
```

```python
pd.set_option('max_columns', None)
pd.set_option('max_rows', None)

df = df.groupby(by=['class_num', 'class_name']).sum().reset_index()
df
```

```python
df.to_excel('./도시락_웹크롤링_데이터현황.xlsx')
```

# background 모델 평가

```python
total = 0
none = 0
correct = 0

with open('./웹데이터test결과.txt', "r", encoding='utf-8') as of:
    lines = of.readlines()
    for line in lines:
        p = re.compile(r'.*pred:.*')
        m = p.findall(line)
        if m:
            p = re.compile(r'.*[락][/](?P<front>.*)(?P<pred>.jpg: pred:)(?P<back>.*)')
            m_label = p.sub("\g<front>", line).strip()
            m_pred = p.sub("\g<back>", line).strip()
            print('label: ', m_label)
            print('pred: ', m_pred)
            total += 1
            if m_pred.strip() == 'None':
                none += 1
            else:
                label = int(m_label.split('_')[2])
                print('label_num: ', label)
                pred = m_pred.split(' ')[-1][:-4]
#                 print(pred.split('e+'))
                if len(pred.split('e+')) == 2:
                    first = float(pred.split('e+')[0])
                    second = int(pred.split('e+')[1])
                    pred_ = int(first*(10**second))
#                     print(label)
                    print('pred_num1: ', pred_)
                    if label == pred_:
                        print('correct',label,pred_)
                        correct += 1
                else:
                    pred_ = int(float(pred.split('e+')[0]))
                    print('pred_num2: ', pred_)
                    if label == pred_:
                        print('correct',label,pred_)
                        correct += 1
            print('=====================')
                        
print('*'*30)
print('결과')
print(f'total_data_count: {total}, bbox_pred: {total-none}, class_correct: {correct}')
```
