```python
import os
import re
import unicodedata
```

# import unicodedata # ㄲㅐㅈㅣㄴㄱㅓㅅ ㅂㅜㅌㅇㅣㄱㅣ

```python
path = '/content/drive/MyDrive/음식인식/multi/'

p = re.compile('[가-힣]*')

box = []
for name in os.listdir(path):
  name = unicodedata.normalize('NFC',name)
  m = p.findall(name)
  for item in m:
    if item:
      box.append(item)

box
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('/content/drive/MyDrive/음식인식/조리식품 학습중인 데이터 .xlsx', sheet_name=1)
df.head()
```

```python
df['multi 건수'] = 0
```

```python
exception = []

for box_name in box:
  # print(f'{box_name}')
  # print(df.loc[df.FOOD_NM==f'{box_name}','multi 건수'])
  if box_name in list(df['FOOD_NM']):
    df.loc[df.FOOD_NM==f'{box_name}','multi 건수'] += 1
  else:
    exception.append(box_name)
```
